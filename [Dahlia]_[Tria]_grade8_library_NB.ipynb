```python
import requests
import json
"""
The 'books' variable is already loaded from the URL in a previous cell.
Therefore, we can directly use the 'books' variable.
"""
data = books
print("JSON file loaded")
ref = db.reference("books")
for book in data["books"]:
  ref.child(str(book["id"])).set(book)
print("Data uploaded successfully!")
```

```python
for key, book in books_data.items():
    details = book["details"]
    available_data = details["available"]
    print(book["title"], "|Available copies:", available_data)
```

```python
total_books = 0
for key, book in books_data.items():
    details = book["details"]
    copies = details["copies"]
    print(book["title"], "| Number of copies:", copies)
    total_books += copies
print("Total number of books in the library:", total_books)
```
